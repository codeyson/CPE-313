Remove characters from the lyrics

```python
import re

newline = lambda x: re.sub(r'\n', ' ', x) # remove \n
alphanumeric = lambda x: re.sub(r'\w*\d\w*', ' ', x) # remove alphanumeric words

corpus = beatles.text.map(newline).map(alphanumeric)
corpus.head()
```

Create a document term matrix using Count Vectorizer with the stop words turned on to English

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words="english")
X = cv.fit_transform(corpus).toarray()

dt = pd.DataFrame(X, columns=cv.get_feature_names_out()).set_index(beatles.song)
dt.head()
```

Imagine lyrics in Count Vectorizer form


```python
imagine = list(dt.loc['Imagine'])
imagine[:20]
```

Define the cosine similarity calculation

```python
from numpy import dot
from numpy.linalg import norm

cosine = lambda v1, v2: dot(v1, v2) / (norm(v1) * norm(v2))
```

Calculate all similarities and sort by the most similar

```python
similarity = [cosine(imagine, song) for song in X]
sorted(list(zip(similarity, beatles.song)), reverse=True)[1:6]
```

Create the TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer

cv_tfidf = TfidfVectorizer(stop_words="english")
X_tfidf = cv_tfidf.fit_transform(corpus).toarray()

dt_tfidf = pd.DataFrame(X_tfidf, columns=cv_tfidf.get_feature_names_out()).set_index(beatles.song)
dt_tfidf.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

cv_tfidf = TfidfVectorizer(stop_words="english")
X_tfidf = cv_tfidf.fit_transform(corpus).toarray()

dt_tfidf = pd.DataFrame(X_tfidf, columns=cv_tfidf.get_feature_names_out()).set_index(beatles.song)
dt_tfidf.head()
```

Calculate all similarities and sort by the most similar using the TF-IDF data

```python
similarity_tfidf = [cosine(imagine, song) for song in X_tfidf]
sorted(list(zip(similarity_tfidf, beatles.song)), reverse=True)[1:6]
```

## Demonstrate how to remove special characters from text

### Tokenize (words)

```python
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

supp['word_tokens'] = supp.Headlines.map(word_tokenize)
supp.head()
```

### Tokenize (words with remove punctuation) / Tokenize (words with remove special characters and numbers)

```python
# Remove special characters
import re
remove_special = lambda x: re.sub(r'[!?£]' ,' ', x.lower())
supp['special'] = supp.Headlines.map(remove_special).map(word_tokenize)


supp.head()

```

```python
# remove numbers
remove_number = lambda x: re.sub(r'[/d!?]' ,' ', x.lower())
supp['remove_num'] = supp.Headlines.map(remove_number).map(word_tokenize)

supp.head()
```

### Tokenize (words with remove punctuation and all lowercase)

```python
# This regular expression removes all punctuation
import string

punc_re = lambda x: re.sub('[%s]' % re.escape(string.punctuation), ' ', x)
supp['Headlines'] = supp.Headlines.map(punc_re)

supp.head()
```

```python
# This regular expression removes all words with numbers in them and also makes all words lowercase
import re

lower_alpha = lambda x: re.sub(r"""\w*\d\w*""", ' ', x.lower())
supp['Headlines'] = supp.Headlines.map(lower_alpha)

supp.head()
```

### Tokenize (words with remove punctuation, all lowercase and remove stop words)

```python
nltk.download('stopwords')
# Remove stop words
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
```

```python
stop_lambda = lambda x: [y for y in x if y not in stop_words]
supp['tokens_stop'] = supp.remove_num.apply(stop_lambda)
supp.head()
```

### Tokenize (words with remove punctuation, all lowercase and  remove stop words, with speech_tagging)

```python
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
# Parts of speech tagging
from nltk.tag import pos_tag

pos_lambda = lambda x: nltk.pos_tag(x)
supp['tokens_pos'] = (supp.tokens_stop.apply(pos_lambda))
supp.head()
```

## Demonstrate how to measure text similarity using distance

```python
press = supp[supp.Press=='Al Jazeera']
press.head()
```

```python
import re

newline = lambda x: re.sub(r'\n', ' ', x) # remove \n
alphanumeric = lambda x: re.sub(r'\w*\d\w*', ' ', x) # remove alphanumeric words

corpus_supp = press.Headlines.map(newline).map(alphanumeric)
corpus_supp.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv_supp = CountVectorizer(stop_words="english")
X_supp = cv_supp.fit_transform(corpus_supp).toarray()

dt_supp = pd.DataFrame(X_supp, columns=cv_supp.get_feature_names_out()).set_index(corpus_supp)
dt_supp.head()
```

```python
selected_headline = corpus_supp.iloc[0]
press_headline_vector = list(dt_supp.loc[selected_headline])
press_headline_vector[:20]
```

```python
from itertools import combinations

pairs_0 = list(combinations(range(len(corpus_supp)), 2)) # all index combos for corpus_supp

supp_headline_pairs = [(corpus_supp.iloc[a_index], corpus_supp.iloc[b_index]) for (a_index, b_index) in pairs_0]
supp_headline_pairs[:5]
```

```python
from numpy import dot
from numpy.linalg import norm

cosine = lambda v1, v2: dot(v1, v2) / (norm(v1) * norm(v2))
```

```python
results = [cosine(X_supp[a_index], X_supp[b_index]) for (a_index, b_index) in pairs_0]
sorted(zip(results, supp_headline_pairs), reverse=True)[:5]
```
